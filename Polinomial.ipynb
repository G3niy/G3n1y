```python
from IPython.display import display, Math
import numpy as np
import copy
import matplotlib
import matplotlib.pyplot as plt

class Polynomial:
  def __init__(self, coef):
    self.coef = coef
    self.normalize()
    self.deg = len(self.coef) - 1
  def normalize(self):
    while len(self.coef) > 1 and self.coef[-1] == 0:
      self.coef = self.coef[:-1]
  def show(self, var="x"):
    display(Math(self.tolatex(var)))
  def __eq__(self, q):
    if len(self.coef) != len(q.coef): 
      return False
    for i in range (len(self.coef)):
      if self.coef[i] != q.coef[i]:
         return False
    return True
  def __ne__(self, q):
    return not(self == q)
  def __pow__(self, n):
    obj = copy.deepcopy(self)
    for i in range(n-1):
      obj *= self
    return obj
  def __neg__(self):
    for i in range(len(self.coef)):
      self.coef[i] *= -1
    return self
  def __add__(self, P):
    if isinstance(P, int):
      P = Polynomial([P])
    n = max(self.deg, P.deg) + 1
    A = (self.coef + [0] * n)[:n]
    B = (P.coef + [0] * n)[:n]
    C = [A[i] + B[i] for i in range(n)]
    return Polynomial(C)
  def __radd__(self, P):
    if isinstance(P, int):
      P = Polynomial([P])
    n = max(self.deg, P.deg) + 1
    A = (self.coef + [0] * n)[:n]
    B = (P.coef + [0] * n)[:n]
    C = [A[i] + B[i] for i in range(n)]
    return Polynomial(C)
  def __iadd__(self, P):
    if isinstance(P, int):
      P = Polynomial([P])
    n = max(self.deg, P.deg) + 1
    A = (self.coef + [0] * n)[:n]
    B = (P.coef + [0] * n)[:n]
    C = [A[i] + B[i] for i in range(n)]
    return Polynomial(C)
  def __sub__(self, P):
    if isinstance(P, int):
      P = Polynomial([P])
    n = max(self.deg, P.deg) + 1
    A = (self.coef + [0] * n)[:n]
    B = (P.coef + [0] * n)[:n]
    C = [A[i] - B[i] for i in range(n)]
    return Polynomial(C)
  def __rsub__(self, P):
    if isinstance(P, int):
      P = Polynomial([P])
    n = max(self.deg, P.deg) + 1
    A = (self.coef + [0] * n)[:n]
    B = (P.coef + [0] * n)[:n]
    C = [A[i] - B[i] for i in range(n)]
    return Polynomial(C)
  def __isub__(self, P):
    if isinstance(P, int):
      P = Polynomial([P])
    n = max(self.deg, P.deg) + 1
    A = (self.coef + [0] * n)[:n]
    B = (P.coef + [0] * n)[:n]
    C = [A[i] - B[i] for i in range(n)]
    return Polynomial(C)
  def __mul__(self, q):
    if isinstance(q, int):
      q = Polynomial([q])
    if isinstance(q, float):
      q = Polynomial([q])
    mass = [0] * (self.deg + q.deg + 1)
    for i in range(len(self.coef)):
      for j in range(len(q.coef)):
        mass[i+j] += self.coef[i] * q.coef[j]
    return Polynomial(mass)
  def __rmul__(self, q):
    if isinstance(q, int):
      q = Polynomial([q])
    if isinstance(q, float):
      q = Polynomial([q])
    mass = [0] * (self.deg + q.deg + 1)
    for i in range(len(self.coef)):
      for j in range(len(q.coef)):
        mass[i+j] += self.coef[i] * q.coef[j]
    return Polynomial(mass)
  def __imul__(self, q):
    if isinstance(q, int):
      q = Polynomial([q])
    if isinstance(q, float):
      q = Polynomial([q])
    mass = [0] * (self.deg + q.deg + 1)
    for i in range(len(self.coef)):
      for j in range(len(q.coef)):
        mass[i+j] += self.coef[i] * q.coef[j]
    return Polynomial(mass)
  def delenie(self, q):
    if self.deg < q.deg:
      return [Polynomial([0]), q]
    massDiv = [0] * (self.deg - q.deg + 1)
    obj = copy.deepcopy(self)
    while(obj.deg >= q.deg):
      massDiv[obj.deg - q.deg] = (obj.coef[-1] / q.coef[-1])
      massHelp = [0] * (obj.deg - q.deg)
      massHelp.append(1)
      obj -= (q * (obj.coef[-1] / q.coef[-1]) * Polynomial(massHelp))
    mod = obj
    div = Polynomial(massDiv)
    return [div, mod]
  def __floordiv__(self, q):
    res = self.delenie(q)
    return res[0]
  def __mod__(self, q):
    res = self.delenie(q)
    return res[1]
  def tolatex(self, var="x"):
    if self.deg == 0:
      return str(self.coef[0])
    s = ""
    for i in range(self.deg, -1, -1):
      if self.coef[i] == 0:
        continue
      # знак
      sign = ""
      if self.coef[i] > 0 and i < self.deg:
        sign = "+"
      elif self.coef[i] < 0:
        sign = "-"
      # value
      v = abs(self.coef[i])
      value = str(v) if (v != 1 or i == 0) else ""
      # degree
      if i == 0:
        degree = ""
      elif i == 1:
        degree = var
      else:
        degree = f"{var}^{{{i}}}"
      s += sign + value + degree
    return s
  def integrate(self, C=0):
    if(self.deg == 0 and self.coef[0] == 0): 
      return C
    mass = [0] * (len(self.coef) + 1)
    for i in range(1, len(mass)):
      mass[i] = self.coef[i-1]/i
    mass[0] = C
    return Polynomial(mass)
  def diff(self, n = 1):
    if self.deg == 0:
      return Polynomial([0])
    obj = copy.deepcopy(self)
    while(n > 0):
      q = []
      for i, c in enumerate(obj.coef):
        if i > 0: 
          q.append(i * c)
      n -= 1
      obj = Polynomial(q)
    return Polynomial(q)
  def __call__(self, x):
    y = 0
    for i in range(self.deg, -1, -1):
      y = y * x + self.coef[i]
    return y
  def plot(self, a, b):
    dx = (b - a) / 100
    X = np.arange(a, b + dx, dx)
    Y = self.__call__(X)
    plt.plot(X, Y)

  

def factor(P):
  # k0 - количество делителей вида x
  k0 = 0
  while k0 < P.deg and P.coef[k0] == 0:
    k0 += 1
  Q = P.coef[k0:] # это будет неразлагаемая часть многочлена 
  X = [] # список делителей-биномов
  stop = False
  while not stop and len(Q) > 1:
    a, b = abs(Q[-1]), abs(Q[0])
    A = factorInteger(a)
    B = factorInteger(b)
    L = makebinomials(A, B)
    stop = True
    for p in L:
      D, R = dividePolynomials(Q, p)
      if R == [0]: # поделилось!
        Q = D
        X.append(p)
        stop = False
        break
  return Q, k0, X

def factorInteger(n):
  return [i for i in range(1, n + 1) if n % i == 0]

def makebinomials(A, B):
  L = []
  for p in A:
    for q in B:
      if get_gcd(p, q) == 1:
        for s in [-1, 1]:
          L.append((s * q, p))
  return L

def get_gcd(n, m):
  while m > 0:
    n, m = m, n % m
  return n

def dividePolynomials(P, Q):
  R, D = P[:], [] # R - остаток от деления P на Q, D - целая часть от этого деления 
  k = len(Q)
  while len(R) >= k:
    if R[-1] % Q[-1] != 0: # целочисленное деление невозможно
      return None, None
    w = R[-1] // Q[-1]
    for i in range(1, k + 1):
      R[-i] -= w * Q[-i]
    R, D = R[:-1], [w] + D
  return D, R

def showFactors(Q, k, X, var = "x"):
  if k == 0 and X == []:
    s = Polynomial(Q).tolatex(var)
  else:
    if len(Q) > 1:
      s = f"({Polynomial(Q).tolatex(var)})"
    elif Q[0] not in [-1, 1]:
      s = str(Q[0])
    elif Q[0] == -1:
      s = "-"
    else:
      s = ""
    X.sort()
    stel = 1
    for x in X:
      if stel > 1:
        stel -= 1
        continue
      else:
        stel = X.count(x)
        if(stel == 1):
          s += f"({Polynomial(x).tolatex(var)})"
        else:
          s += f"({Polynomial(x).tolatex(var)})^{{{stel}}}"
    if k == 1:
      s += var
    elif k > 1:
      s += f"{var}^{{{k}}}"
  display(Math(s))
```
